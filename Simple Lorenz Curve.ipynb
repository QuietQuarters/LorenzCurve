# Lorenz Curve

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from scipy.integrate import simps
from numpy import trapz
```

```python
income_input=input('Insert the level of national income in order by quintiles, seperated by a space:  ')
income_input = income_input.split(' ')
income_input = list(map(int, income_input))

def inc_cum(x):
    cum20 = x[0]/(sum(x)) *100
    cum40 = (x[0]+x[1])/(sum(x)) *100
    cum60 = (x[0]+x[1]+x[2])/(sum(x)) *100
    cum80 = (x[0]+x[1]+x[2]+x[3])/(sum(x)) *100
    listinc=[0, cum20, cum40, cum60, cum80, 100]
    return pd.DataFrame(listinc, columns=['Cumulative Frequency of Income'])

def inc_cum_list(x):
    cum20 = x[0]/(sum(x)) *100
    cum40 = (x[0]+x[1])/(sum(x)) *100
    cum60 = (x[0]+x[1]+x[2])/(sum(x)) *100
    cum80 = (x[0]+x[1]+x[2]+x[3])/(sum(x)) *100
    listinc=[0, cum20, cum40, cum60, cum80, 100]
    return listinc

df1 = inc_cum(income_input)
df1['Quintile'] = [0, 1, 2, 3, 4, 5]
df1=df1.reindex(columns=['Quintile',  'Cumulative Frequency of Income'])

lst1=[1,1,1,1,1]
dfeq = inc_cum(lst1)
dfeq['Quintile'] = [0, 1, 2, 3, 4, 5]


popcum = {'Quintile': [0, 1, 2, 3, 4, 5], 'Cumulative Frequency of population': [0, 20, 40, 60, 80, 100]}
dfpop=pd.DataFrame(popcum)

df3=dfpop.merge(df1, on='Quintile')
df_perfequality=dfeq.merge(dfpop, on='Quintile')

sns.set_style('whitegrid')
ax=sns.lineplot(data=df_perfequality, x='Cumulative Frequency of population', y='Cumulative Frequency of Income')
g=sns.lineplot(data=df3, ax=ax, x='Cumulative Frequency of population', y='Cumulative Frequency of Income')
g.set_title('Lorenz Curve')
plt.show()


# The y values.  A numpy array is used here,
# but a python list could also be used.
y = np.array(inc_cum_list(income_input))
y_perf = np.array([inc_cum_list([1,1,1,1,1])])

# Compute the area using the composite trapezoidal rule.
area_curve = trapz(y, dx=20)
area_perft = trapz(y_perf, dx=20)
Gini_coeff = (area_perft-area_curve)/area_perft
print('Gini coefficient is', Gini_coeff)
```
